## Unit impulse function

```python
import numpy as np
import matplotlib.pyplot as plt
N=5
n=np.arange(-N,N+1)
x=np.zeros(n.shape)
x[N]=1
plt.stem(n,x)
plt.ylabel('$\delta[n]$')
plt.xlabel('n')
plt.title('x[n]= $\delta[n]$')
```

## unit step function

```python
import numpy as np
import matplotlib.pyplot as plt
N=5
n=np.arange(-N,N+1)
x[N:]=1
plt.grid(True)
plt.stem(n,x)
plt.ylabel('u[n]')
plt.xlabel('n')
plt.title('x[n]= u[n]')
```

## exponential function

```python
import numpy as np
import matplotlib.pyplot as plt
N=3
n=np.arange(-N,15)
x=np.power(0.9,n)
x[:N]=0
plt.grid(True)
plt.stem(n,x)
plt.ylabel('x[n]')
plt.xlabel('n')
plt.title('x[n]= $(0.9)^nu[n]$')
```

## Sinusodial sequence

```python
import numpy as np
import matplotlib.pyplot as plt
N=15
n=np.arange(-N,N+1)
omega=np.pi*2/10
x=np.sin(omega*n)
plt.grid(True)
plt.stem(n,x)
plt.plot(n,x,'r')
plt.ylabel('x[n]')
plt.xlabel('n')
plt.title(r'$x[n]=\ cos(2\pi n/10)$')
```
